```python
#separate the features from the labels
y = reg_data['total_claim_amount']
X = reg_data.drop(['total_claim_amount'], axis=1)
```

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = np.object)
```

```python
MinMaxtransformer = MinMaxScaler().fit(X_num)
x_normalized = MinMaxtransformer.transform(X_num)
x_normalized.shape
x_normalized = pd.DataFrame(x_normalized,columns=X_num.columns)
x_normalized.head(2)
```

```python
Standardtransformer = StandardScaler().fit(X_num)
x_standardized = Standardtransformer.transform(X_num)
x_standardized.shape
x_standardized = pd.DataFrame(x_standardized,columns=X_num.columns)
x_standardized.head(2)
```

```python
label_encoded = LabelEncoder().fit(X_cat).transform(X_cat) 
label_encoded = pd.DataFrame(label_encoded,columns=X_cat.columns)
label_encoded
type(label_encoded)
#label_encoded.value_counts()
```

```python
X = pd.concat([x_normalized, label_encoded], axis=1)
```

```python
X.head()
```

```python
#y = reg_data['total_claim_amount']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1000)
```
