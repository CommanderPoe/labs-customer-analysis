```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('file1.csv')
df2 = pd.read_csv('file2.csv')
df3 = pd.read_csv('file3.csv')
```

```python
# Turn all columns headers lowercase
df1.columns= df1.columns.str.lower()
#df1.head()
#df1.head()
df2.columns= df2.columns.str.lower()
#df2.head()
df3.columns= df3.columns.str.lower()
#df3.head()
#df1.columns

#df1.head(3)
#df2.head(3)
#df3.head(3)
```

```python
df1= df1.rename(columns= {'st': 'state'})
df2= df2.rename(columns= {'st': 'state'})
df3= df3.rename(columns= {'st': 'state'})

#...(columns = {'test':'TEST'}, inplace = True)

df1.head(1)
df2.head(1)
df3.head(1)

df1.columns
```

```python
df2 = df2[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]

df3= df3[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]

df1.head()
df2.head()
df3.head()
```

```python
final = pd.concat([df1,df2,df3], axis=0)
final.sample(10)
final.shape
```

```python
final.info()
final.dtypes
#final[['income', 'monthly premium auto', 'total claim amount']]
final._get_numeric_data()
final.select_dtypes('object')
```

```python
#Deleting the edu column
final.drop(columns= 'education', inplace= True, axis= 1)
```

```python
final.drop(columns= 'number of open complaints', inplace= True, axis= 1)
```

```python
final.head()
```

```python
"""Correct the values in the column customer lifetime value. 
They are given as a percent, so multiply them by 100 and change dtype to numerical type"""

final['customer lifetime value']= final['customer lifetime value'].str.replace('%','')
```

```python
final['customer lifetime value']= pd.to_numeric(final['customer lifetime value'], errors= 'coerce')
final.dtypes

final['customer lifetime value']= final['customer lifetime value']*100

#round(final['customer lifetime value'], 1)
final.head(5)
```

```python
# Check for duplicate rows in the data and remove if any
final.duplicated().value_counts()
#final= final.duplicated(keep='first')

#final.drop_duplicates(inplace= True)
#final.info()
final.head()
final.shape
```

```python
final.drop_duplicates(inplace= True)
final.shape
```

```python
#Filter out the data for customers who have an income of 0 or less.
poor_people= final[final['income']<= 0]
poor_people.describe()
poor_people.shape
```
